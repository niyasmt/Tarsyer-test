```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('Task_3.jpg', 0)
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
cv.imwrite('Task_3_simple.jpg', thresh1)

img = cv.medianBlur(img,5)
th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)
cv.imwrite('Task_3_adaptive.jpg', th3)

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.subplot(1, 3, 2)
plt.imshow(thresh1)
plt.subplot(1, 3, 1)
plt.imshow(th3)
```
